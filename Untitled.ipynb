```python
tweets = []
```

```python
for t in df.tweet:
    tweets.append(t)
```

```python
def add_to_arr(arr, func):
    temp = []
    for a in arr:
        temp.append(func(a))
    return temp
```

```python
def add_col(df ,arr, name):
    df[name] = arr
    return df
```

# 1:

```python
def count_words(txt):
    return len(txt.split())
```

```python
num_of_big_words_arr = add_to_arr(tweets, count_words)
```

```python
add_col(df, num_of_big_words_arr, 'Count of words')
```
